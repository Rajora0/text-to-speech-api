```python
import json
import requests
from IPython.display import Audio
```

```python
# Define a URL da API
url = 'http://localhost:5000/synthesize'

# Define o payload com o speaker e o texto
payload = {
    "speaker": "0",  # Substitua pelo ID do speaker que deseja usar
    "text": "Hello, my dog is cooler than you!"
}
```

```python
# Envia a requisição POST para a API
response = requests.post(url, json=payload)

# Verifica o status da resposta
if response.status_code == 200:
    # Salva o arquivo WAV retornado
    with open("response_speech.wav", "wb") as f:
        f.write(response.content)
    print("Audio saved as response_speech.wav")
    
    # Exibe os metadados
    metadata = response.headers.get('metadata', '')
    print("Metadata:", metadata)
else:
    print(f"Request failed with status code {response.status_code}")
    print("Response:", response.text)
```

```python
# Play the synthesized speech
Audio('response_speech.wav', rate=16000)
```
